```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# File to Load (Remember to change these)
city_data_to_load = os.path.join(".","data","city_data.csv")
ride_data_to_load = os.path.join(".","data","ride_data.csv")

# Read the City and Ride Data
city_data = pd.DataFrame(pd.read_csv(city_data_to_load))
ride_data = pd.DataFrame(pd.read_csv(ride_data_to_load))

# Combine the data into a single dataset
df = ride_data.merge(city_data, on="city")

# Display the data table for preview
df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Generate X Coords
farecount = pd.DataFrame(df.groupby(['city','type'])['fare'].count())
farecount = farecount.reset_index()

x_urban = farecount.loc[farecount['type']=='Urban',:]

x_suburban = farecount.loc[farecount['type']=='Suburban',:]

x_rural = farecount.loc[farecount['type']=='Rural',:]


# Generate Y Coords
avgfare = pd.DataFrame(df.groupby(['city','type']).mean())
avgfare = avgfare.reset_index()

y_urban = avgfare.loc[avgfare['type']=='Urban',:]
y_urban = y_urban.drop(columns='ride_id')

y_suburban = avgfare.loc[avgfare['type']=='Suburban',:]
y_suburban = y_suburban.drop(columns='ride_id')

y_rural = avgfare.loc[avgfare['type']=='Rural',:]
y_rural = y_rural.drop(columns='ride_id')

# Build the scatter plots for each city types
urban_size = [x * 10 for x in y_urban['driver_count']]
suburban_size = [x * 10 for x in y_suburban['driver_count']]
rural_size = [x * 10 for x in y_rural['driver_count']]

plt.scatter(x_urban['fare'], y_urban['fare'], s=urban_size, edgecolors="black", facecolors='coral', alpha=0.6, label='Urban')
plt.scatter(x_suburban['fare'], y_suburban['fare'], s=suburban_size, edgecolors="black", facecolors='skyblue', alpha=0.6, label='Suburban')
plt.scatter(x_rural['fare'], y_rural['fare'], s=rural_size, edgecolors="black", facecolors='gold', alpha=0.6, label='Rural')

# Incorporate the other graph properties
plt.grid()
plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare (USD)')

# Create a legend
legend = plt.legend(loc='upper right', title='City Types', markerscale=1)
legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
plt.text(43,30, "Note: Circle size\ncorrelates with\ndriver count per city.", bbox=dict(facecolor='white', alpha=0.5))

# Save Figure
plt.savefig('images/hw_jchung_pyber_plot.png', bbox_inches="tight")

# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fare_total = pd.DataFrame(df.groupby(['type'])['fare'].sum())
fare_total = fare_total.reset_index()

labels = list(fare_total['type'])
sizes = list(fare_total['fare'])
colors = ["gold", "lightskyblue", "lightcoral"]
explode = [0.0, 0.0, 0.1]

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True)
plt.title('% of Total Fares by City Type')

# Save Figure
plt.savefig('images/hw_jchung_pyber_pie1.png', bbox_inches="tight")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Type Percents
fare_type = pd.DataFrame(df.groupby(['type'])['fare'].count())
fare_type = fare_type.reset_index()

labels = list(fare_type['type'])
sizes = list(fare_type['fare'])
colors = ["gold", "lightskyblue", "lightcoral"]
explode = [0.0, 0.0, 0.1]

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True)
plt.title('% of Total Rides by City Type')

# Save Figure
plt.savefig('images/hw_jchung_pyber_pie2.png', bbox_inches="tight")
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Type Percents
sizes = []

rural_drivers = city_data.loc[city_data['type']=='Rural',:]
sizes.append(rural_drivers['driver_count'].sum())
sizes

suburban_drivers = city_data.loc[city_data['type']=='Suburban',:]
sizes.append(suburban_drivers['driver_count'].sum())

urban_drivers = city_data.loc[city_data['type']=='Urban',:]
sizes.append(urban_drivers['driver_count'].sum())

colors = ["gold", "lightskyblue", "lightcoral"]
explode = [0.0, 0.0, 0.1]

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True)
plt.title('% of Total Drivers by City Type')

# Save Figure
plt.savefig('images/hw_jchung_pyber_pie3.png', bbox_inches="tight")
plt.show()
```
